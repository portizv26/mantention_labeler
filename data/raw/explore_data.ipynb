```python
import pandas as pd
import os
from tqdm import tqdm

os.chdir('../')
```

```python
df = pd.read_excel('data/raw/maintenance_data_raw.xlsx')

df.Fecha_Actual = pd.to_datetime(df.Fecha_Actual, format='%d/%m/%Y %H:%M:%S')

df.loc[:, 'weekYear'] = df.Fecha_Actual.dt.strftime('%Y-%U')

for wy in tqdm(df.weekYear.unique(), desc="Processing weeks"):
    df_aux = df[df.weekYear == wy].copy()
    df_aux.drop(columns=['weekYear'], inplace=True)
    df_aux.to_excel(f'data/to_process/maintenance_data_{wy}.xlsx', index=False)
```
